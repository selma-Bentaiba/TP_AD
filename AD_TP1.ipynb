## Q1 :

```python
import numpy as np  # we use NumPy for matrix operations

# ---------------------------
# Step 1: Define the Data Matrix X
# ---------------------------
# The matrix represents student scores in 5 different subjects
# Rows (E1 to E9) represent students
# Columns represent subjects: Multimedia, Maths, System, Network, Other

X = np.array([
    [6, 6, 5, 5.5 , 8],
    [8, 8, 8, 8, 9],
    [6, 7, 11, 9.5, 11],
    [14.5, 14.5, 15.5, 15, 8],
    [14, 14, 12, 12, 10],
    [11, 10, 5.5, 7, 13],
    [5.5, 7, 14, 11.5, 10],
    [13, 12.5, 8.5, 9.5, 12],
    [9, 9.5, 12.5, 12, 18]
])

# Print the original matrix
print("Original Matrix X:")
print(X)
```

```python
# Step 2: Compute the Transpose of Matrix X
# ---------------------------
# Transposing a matrix means swapping rows and columns.
# In the transposed matrix X^T:
#   - Rows become columns
#   - Columns become rows
#
# Example:
# X (Original Matrix)         X^T (Transposed Matrix)
# [a, b, c]                   [a, d, g]
# [d, e, f]       -->         [b, e, h]
# [g, h, i]                   [c, f, i]

X_t = X.T  # Using NumPy's built-in transpose function

# Print the transposed matrix
print("\nTransposed Matrix X^T:")
print(X_t)
```

## Q2 :

```python
# Step 3: Extract the List of Individuals (Students)
# ---------------------------
# Since each row represents a student, we can generate a list of student names (E1 to E9)
# we can use Python's range function to create labels dynamically



#X.shape[0] gives the number of rows = the number of students (9 in this case).
# range(X.shape[0]) = range(9) â†’ [0, 1, 2, 3, 4, 5, 6, 7, 8]
#str(i+1) -> [1, 2, ..9]


individuals = ["E" + str(i+1) for i in range(X.shape[0])]


# Print the list of individuals
print("\nList of Individuals (Students):")
print(individuals)
```

## Q3 :

```python
# Extracting each column (variable) into separate vectors
#-------------------------------------------------------
# X[:, i] means:
# : selects all rows.
# i selects the i-th column (starting from 0 in Python)





multimedia = X[:, 0]  # First column (Multimedia)
maths = X[:, 1]       # Second column (Maths)
system = X[:, 2]      # Third column (System)
network = X[:, 3]     # Fourth column (Network)
other = X[:, 4]       # Fifth column (Other)


# Display the extracted vectors

print("Multimedia scores:", multimedia)
print("Maths scores:", maths)
print("System scores:", system)
print("Network scores:", network)
print("Other scores:", other)
```

## Q4

```python
# Retrieve the rows (individuals) corresponding to students 4, 5, and 7 from the data matrix X


# Extracting individuals 4, 5, and 7
individual_4 = X[3, :]  # Row at index 3 (E4)
individual_5 = X[4, :]  # Row at index 4 (E5)
individual_7 = X[6, :]  # Row at index 6 (E7)

# Display results
print("E4 (4th Individual):", individual_4)
print("E5 (5th Individual):", individual_5)
print("E7 (7th Individual):", individual_7)

```

## Q5 :

```python
# Creating a Table with Mean (moyenne), Variance, and Standard Deviation (Ã©cart type) for Each Variable / column (module)

means = np.mean(X, axis=0)   # Mean of each column
variances = np.var(X, axis=0, ddof=0)  # Variance of each column
std_devs = np.std(X, axis=0, ddof=0)   # Standard deviation of each column

# Round values to 4 decimal places
means = np.round(means, 4)
variances = np.round(variances, 4)
std_devs = np.round(std_devs, 4)

```

```python
import pandas as pd # manipulate and analyze data

# we create a DataFrame for better visualization

columns = ["Multimedia", "Maths", "System", "Network", "Other"]
stats_table = pd.DataFrame({
    "Mean": means,
    "Variance": variances,
    "Std Dev": std_devs
}, index=columns)

# Display table
print(stats_table)
```

## Q6 :

```python
# compute the average scores of each student across all modules to determine the average individual

# Compute the average score per student
average_individual = np.mean(X, axis=1)  # Mean of each row (student)

# Round values to 4 decimal places
average_individual = np.round(average_individual, 4)

# Display results
for i, avg in enumerate(average_individual):
    print(f"Student E{i+1}: {avg}")
```

## Q7:

```python
#Write a Function to Compute the Variance of the 5 Variables in the Given Matrix (Rounded to 4 Decimal Places)

# Function to compute variance for each module (column)

def compute_variance(matrix):
    variance_values = np.var(matrix, axis=0)  # Compute variance column-wise
    variance_values = np.round(variance_values, 4)  # Round to 4 decimal places
    return variance_values

# Compute and display variance
variances = compute_variance(X)

# Display results
module_names = ["Multimedia", "Maths", "System", "Network", "Other"]
for i, var in enumerate(variances):
    print(f"Variance of {module_names[i]}: {var}")
```

## Q8 :

```python
# Compute All Possible Covariances Between Variables and Display the Results in a Matrix ð‘‰

# Compute the covariance matrix
V = np.cov(X, rowvar=False)  # rowvar=False ensures we compute column-wise covariance
V = np.round(V, 4)  # Round to 4 decimal places

# Display the covariance matrix
print("Covariance Matrix (V):\n", V)
```

## Q12 :

 we need to graphically represent scatter plots (nuage de points) for the following variable pairs in ð‘…Â² : (X1,X2),(X2,X4),(X3,X4).

```python
import numpy as np
import matplotlib.pyplot as plt

# Extract variable pairs
X1, X5 = X[:, 0], X[:, 4]  # (X1, X5)
X2, X4 = X[:, 1], X[:, 3]  # (X2, X4)
X3, X4 = X[:, 2], X[:, 3]  # (X3, X4)

# Create figure with 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Scatter plot for (X1, X5)
axs[0].scatter(X1, X5, color='red', edgecolors='black')
axs[0].set_title("Scatter Plot (X1, X5)")
axs[0].set_xlabel("X1 (Multimedia)")
axs[0].set_ylabel("X5 (Other)")

# Scatter plot for (X2, X4)
axs[1].scatter(X2, X4, color='blue', edgecolors='black')
axs[1].set_title("Scatter Plot (X2, X4)")
axs[1].set_xlabel("X2 (Maths)")
axs[1].set_ylabel("X4 (Networks)")

# Scatter plot for (X3, X4)
axs[2].scatter(X3, X4, color='green', edgecolors='black')
axs[2].set_title("Scatter Plot (X3, X4)")
axs[2].set_xlabel("X3 (Systems)")
axs[2].set_ylabel("X4 (Networks)")

# Display the plots
plt.tight_layout()
plt.show()
```
